Merge customer codes with their categories:

```python
joined = receipts.merge(customers, how='inner', on='KCustomer')#, lsuffix='', rsuffix='_right')
joined
```

Quality and easyness stats for ach category:

```python
import json

for cat in categories:
    subsample = joined[joined['Description'] == cat]
    print(cat, len(subsample))
    cat_quals, q_counts = np.unique(subsample['QualClass'], return_counts=True)
    cat_easys, e_counts = np.unique(subsample['EasyClass'], return_counts=True)
    q_indices = q_counts.argsort()[::-1]
    e_indices = e_counts.argsort()[::-1]
    for idx in q_indices:
        print(cat_quals[idx], round(q_counts[idx] / len(subsample), 2))
    for idx in e_indices:
        print(cat_easys[idx], round(e_counts[idx] / len(subsample), 2))
    print('---------------------------')
```

### K Group API

Get info about one particular store:

```python
def get_store_coords(store_ids=['K171'], key='174385768bde498fa5ac43a4521143d8'):
    import http.client, urllib.request, urllib.parse, urllib.error, base64

    headers = {
        # Request headers
        'Content-Type': 'application/json',
        'Ocp-Apim-Subscription-Key': key,
    }

    params = urllib.parse.urlencode({})
    
    body_dict = {'filters': 
                     {
                         'ids': store_ids
                     }
                }
    body = json.dumps(body_dict)
    print(body)

    try:
        conn = http.client.HTTPSConnection('kesko.azure-api.net')
        conn.request("POST", "/v1/search/stores?%s" % params, body, headers)
        response = conn.getresponse()
        data = response.read()
        conn.close()
        return json.loads(data)
    except Exception as e:
        print("[Errno {0}] {1}".format(e.errno, e.strerror))
```

Function for saving info about all stores to CSV file:

```python
def get_store_coords_all(store_ids=['K171'], key='174385768bde498fa5ac43a4521143d8'):
    import http.client, urllib.request, urllib.parse, urllib.error, base64

    headers = {
        # Request headers
        'Content-Type': 'application/json',
        'Ocp-Apim-Subscription-Key': key,
    }

    params = urllib.parse.urlencode({})
    
    body_dict = {"filters": 
                 {
                     "locationDistance": 
                     {
                         "location": 
                         {
                            "lon": 24,
                            "lat": 60
                         },
                         "distance": 50
                     }
                 },
                 "view":
                 {
                    "limit": 100
                 }
                }
    body = json.dumps(body_dict)
    print(body)

    try:
        conn = http.client.HTTPSConnection('kesko.azure-api.net')
        conn.request("POST", "/v1/search/stores?%s" % params, body, headers)
        response = conn.getresponse()
        data = response.read()
        conn.close()
        return json.loads(data)
    except Exception as e:
        print("[Errno {0}] {1}".format(e.errno, e.strerror))
```

```python
coords_all = get_store_coords_all()
coords_all
```

```python
def save_markets():
    stores = get_store_coords_all()
    df = []
    for store_info in stores['results']:
        try:
            row = {}
            row['name'] = store_info['Name']
            row['long'] = store_info['Coordinate']['Longitude']
            row['lat'] = store_info['Coordinate']['Latitude']
            row['address'] = store_info['Address']
            row['phoneNumber'] = store_info['PhoneNumber']
            row['opens'] = store_info['OpeningHoursSpecification'][0]['opens']
            row['closes'] = store_info['OpeningHoursSpecification'][0]['closes']
            row['dayOfWeek'] = store_info['OpeningHoursSpecification'][0]['dayOfWeek'].split("/")[-1]
            df.append(row)
        except:
            continue
    return pd.DataFrame(data=df)
```

```python
df = save_markets()
df
```
